### Load the data


```python
df = pd.read_csv("survey-data.csv")
```

### Display the first few rows of the data to understand its structure


```python
df.head()
```

### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
# Group by Age and calculate median for the two satisfaction scores
age_satisfaction = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot stacked bar chart
age_satisfaction.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Median Job Satisfaction Scores by Age Group")
plt.xlabel("Age")
plt.ylabel("Median Job Satisfaction Points")
plt.xticks(rotation=45)
plt.legend(title='JobSatPoints')
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
# Group by Employment status and get median satisfaction scores
employment_satisfaction = df.groupby('EmploymentStatus')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot stacked bar chart
employment_satisfaction.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Median Job Satisfaction by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Median Job Satisfaction Points")
plt.xticks(rotation=45)
plt.legend(title='JobSatPoints')
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
# Filter respondents aged 30-35
age_30_35 = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Median compensation and satisfaction
median_30_35 = age_30_35.groupby('Age')[['ConvertedCompYearly','JobSatPoints_6']].median()

median_30_35.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title("Median Compensation and Job Satisfaction (Age 30-35)")
plt.xlabel("Age")
plt.ylabel("Median Value")
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
# Median compensation and JobSatPoints_6 by Age
median_by_age = df.groupby('Age')[['ConvertedCompYearly','JobSatPoints_6']].median()

median_by_age.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Median Compensation and Job Satisfaction Across Age Groups")
plt.xlabel("Age")
plt.ylabel("Median Value")
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
# Create a pivot table: Age vs DatabaseWantToWorkWith
db_age = df.pivot_table(index='Age', columns='DatabaseWantToWorkWith', aggfunc='size', fill_value=0)

# Only keep top 5 databases overall
top5_db = df['DatabaseWantToWorkWith'].value_counts().head(5).index
db_age_top5 = db_age[top5_db]

# Plot stacked chart
db_age_top5.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Top 5 Desired Databases by Age Group")
plt.xlabel("Age")
plt.ylabel("Number of Respondents")
plt.legend(title='Database')
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
# Pivot table: JobSatPoints_6 vs EmploymentStatus
employment_js = df.pivot_table(index='JobSatPoints_6', columns='EmploymentStatus', aggfunc='size', fill_value=0)

employment_js.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Employment Type Distribution by Job Satisfaction")
plt.xlabel("Job Satisfaction Points")
plt.ylabel("Number of Respondents")
plt.legend(title='Employment Status')
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
# Pivot: Age vs LanguageAdmired
lang_age = df.pivot_table(index='Age', columns='LanguageAdmired', aggfunc='size', fill_value=0)

# Top 5 admired languages
top5_lang = df['LanguageAdmired'].value_counts().head(5).index
lang_age_top5 = lang_age[top5_lang]

lang_age_top5.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Top 5 Admired Programming Languages by Age Group")
plt.xlabel("Age")
plt.ylabel("Number of Respondents")
plt.legend(title='Language')
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
# Pivot: EmploymentStatus vs PlatformAdmired
platform_emp = df.pivot_table(index='EmploymentStatus', columns='PlatformAdmired', aggfunc='size', fill_value=0)

# Top 5 admired platforms
top5_platforms = df['PlatformAdmired'].value_counts().head(5).index
platform_emp_top5 = platform_emp[top5_platforms]

platform_emp_top5.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Top 5 Admired Platforms by Employment Type")
plt.xlabel("Employment Status")
plt.ylabel("Number of Respondents")
plt.legend(title='Platform')
plt.show()
```
